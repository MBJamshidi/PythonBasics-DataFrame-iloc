```python
#Selection by label
#iloc
#pandas.DataFrame.iloc
#Purely integer-location based indexing for selection by position.
#.iloc[] is primarily integer position based (from 0 to length-1 of the axis)
#but may also be used with a boolean array.
#Allowed inputs are:
#    @An integer, e.g. 5.
#    @A list or array of integers, e.g. [4, 3, 0].
#    @A slice object with ints, e.g. 1:7.
#    @A boolean array.
#    @A callable function with one argument (the calling Series or DataFrame)

import pandas as pd

Dictionary1 = [{'U1': 1, 'U2': 2, 'U3': 3, 'U4': 4, 'U5': 5, 'U6': 6, 'U7': 7},
          {'U1': 100, 'U2': 200, 'U3': 300, 'U4': 400, 'U5': 500,'U6': 600,'U7': 700},
          {'U1': 1000, 'U2': 2000, 'U3': 3000, 'U4': 4000, 'U5': 5000,'U6': 6000,'U7': 7000}]
df2 = pd.DataFrame(Dictionary1)

#Indexing just the rows
#With a scalar integer[]
print("\n indexing just the rows")
print("\n With a scalar integer[0]:")
df3=df2.iloc[0]
print(df3)

print("\n With a scalar integer[2]:")
df3=df2.iloc[2]
print(df3)

print("\n With a scalar integer[1]:")
df3=df2.iloc[1]
print(df3)

#With a list of integers
print("\n With a list of integers[[0]]:")
df3=df2.iloc[[0]]
print(df3)

print("\n With a list of integers[[2]]:")
df3=df2.iloc[[2]]
print(df3)

print("\n With a list of integers[[1,2]]:")
df3=df2.iloc[[1,2]]
print(df3)

#With a slice object
print("\n With a slice object[:3]:")
df3=df2.iloc[:3]
print(df3)

print("\n With a slice object[0:1]:")
df3=df2.iloc[0:1]
print(df3)

print("\n With a slice object[0:2]:")
df3=df2.iloc[0:2]
print(df3)


#With a boolean mask the same length as the index
print("\n With a boolean mask the same length as the index[True,True, False]:")
df3=df2.iloc[[True,True, False]]
print(df3)

```

```python
#Indexing both axes
import pandas as pd

Dictionary1 = [{'U1': 1, 'U2': 2, 'U3': 3, 'U4': 4, 'U5': 5, 'U6': 6, 'U7': 7},
          {'U1': 100, 'U2': 200, 'U3': 300, 'U4': 400, 'U5': 500,'U6': 600,'U7': 700},
          {'U1': 1000, 'U2': 2000, 'U3': 3000, 'U4': 4000, 'U5': 5000,'U6': 6000,'U7': 7000}]
df2 = pd.DataFrame(Dictionary1)

#You can mix the indexer types for the index and columns. Use : to select the entire axis
#Indexing both axes
#With a scalar integer[]
print("\n indexing both axes")
print("\n With a scalar integer[0,1]:")
df3=df2.iloc[0,1]
print(df3)

#With lists of integers
print("\n indexing both axes")
print("\n With lists of integers[[0,1][1,3]]:")
df3=df2.iloc[[0,1],[1,3]]
print(df3)

#With slice objects
print("\n indexing both axes")
print("\n With slice objects [0:3, 0:3]:")
df3=df2.iloc [0:3, 0:3]
print(df3)


#With a boolean array whose length matches the columns
print("\n indexing both axes")
print("\n With a boolean array whose length matches the columns [:, [True, False, True, False, True, True, True]]:")
df3=df2.iloc [:, [True, False, True, False, True, True, True]]
print(df3)

print("\n indexing both axes")
print("\n With a boolean array [[False,True, False], [False, False, True, False, True, True, True]]:")
df3=df2.iloc [[False,True, False], [False, False, True, False, True, True, True]]
print(df3)
```
